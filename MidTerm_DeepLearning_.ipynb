```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
# Step 1: Load and preprocess the data
fashion_mnist = tf.keras.datasets.fashion_mnist

(train_images, train_labels), _= fashion_mnist.load_data()
# Reshape the train_images array from 3D to 2D.
train_images = train_images.reshape((60000, 28*28))
# Convert train_images data type to float32 and normalize pixel values by dividing by 255.
train_images = train_images.astype("float32") / 255
```

```python
import keras
```

```python
import numpy as np
np.unique(train_labels)
```

# Model 1

```python
model = keras.Sequential([
    Dense(16, activation='relu'),
    Dense(16, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer = 'rmsprop',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
history =model.fit(train_images, train_labels, epochs=20, batch_size=128, validation_split=0.2)
```

```python
def plot_history(history):
    history_dict = history.history  # Extracting the history dictionary from the input history object
    loss_values = history_dict['loss']  # Extracting training loss values
    val_loss_values = history_dict['val_loss']  # Extracting validation loss values
    acc_values = history_dict['accuracy']  # Extracting training accuracy values
    val_acc_values = history_dict['val_accuracy']  # Extracting validation accuracy values

    epochs = range(1, len(loss_values) + 1)  # Generating a range of epochs for x-axis

    plt.figure(figsize=(12, 4))  # Setting figure size

    # Plotting training and validation loss
    plt.subplot(1, 2, 1)  # Creating subplot for loss
    plt.plot(epochs, loss_values, 'bo', label='Training loss')  # Plotting training loss
    plt.plot(epochs, val_loss_values, 'b', label='Validation loss')  # Plotting validation loss
    plt.title('Training and validation loss')  # Setting title
    plt.xlabel('Epochs')  # Setting x-axis label
    plt.ylabel('Loss')  # Setting y-axis label
    plt.legend()  # Showing legend

    # Plotting training and validation accuracy
    plt.subplot(1, 2, 2)  # Creating subplot for accuracy
    plt.plot(epochs, acc_values, 'bo', label='Training acc')  # Plotting training accuracy
    plt.plot(epochs, val_acc_values, 'b', label='Validation acc')  # Plotting validation accuracy
    plt.title('Training and validation accuracy')  # Setting title
    plt.xlabel('Epochs')  # Setting x-axis label
    plt.ylabel('Accuracy')  # Setting y-axis label
    plt.legend()  # Showing legend

    plt.show()  # Displaying the plot

plot_history(history)  # Calling the plot_history function with the history1 object
```

#Model 2

```python
model2 = keras.Sequential([
    Dense(16, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
model2.compile(optimizer = 'rmsprop',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
history2 =model2.fit(train_images, train_labels, epochs=20, batch_size=128, validation_split=0.2)
```

```python
plot_history(history2)
```

#Model 3

```python
model3 = keras.Sequential([
    Dense(32, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
model3.compile(optimizer = 'rmsprop',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
history3 =model3.fit(train_images, train_labels, epochs=20, batch_size=128, validation_split=0.2)
```

```python
plot_history(history3)
```

**Model 1 with 16 hidden layer each is the best option among the three. It achieved a good balance between training speed, stability, and accuracy. The test accuracy of 86% is quite good for the MNIST fashion dataset, indicating effective generalization. Although Model 3  with 32 hidden layer achieved the highest validation accuracy, its test accuracy did not significantly surpass that of Model 1.**
